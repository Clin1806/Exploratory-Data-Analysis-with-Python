[Statistical Analysis, Hypothesis Testing & A/B Testing](https://)

```python
df.info()
```

```python
bins = [0, 3, 6, 9]
labels = ['Low', 'Medium', 'High']

df['Air_Pollution_range'] = pd.cut(df['Air Pollution'], bins=bins, labels=labels, right=False)
```

```python
# Check the new column
df[['Patient Id', 'Air Pollution', 'Air_Pollution_range']].head()
```

```python
# Box plot to detect outliers in Air Pollution
plt.figure(figsize=(8, 5))
sns.boxplot(x=df['Air Pollution'], color='lightblue')
plt.title("Outlier Detection: Air Pollution")
plt.xlabel("Air Pollution")
plt.show()
```

```python
# Calculate Wheezing_rates per Air_Pollution category
Wheezing_rates = df.groupby('Air_Pollution_range')['Wheezing'].mean() * 100
print(Wheezing_rates)
```

```python
# Visualizing Wheezing_rates
plt.figure(figsize=(8, 5))
sns.barplot(x=Wheezing_rates.index, y=Wheezing_rates.values, palette='coolwarm')
plt.title("Wheezing_rates by Air_Pollution_range")
plt.xlabel("Air_Pollution_range")
plt.ylabel("Wheezing_rates (%)")
plt.show()
```

[ A/B Testing](https://)

```python

# Let's create two artificial groups: Low Exposure vs. High Exposure

low_exposure = df[df['Air_Pollution_range'] == 'Low']
high_exposure = df[df['Air_Pollution_range'] == 'High']

# Display the size of each group
print(f"Low Exposure Group Size: {len(low_exposure)}")
print(f"High Exposure Group Size: {len(high_exposure)}")
```

```python
from scipy.stats import ttest_ind

# # Perform a t-test on 'Wheezing' values between low and high exposure groups.

t_stat, p_value = ttest_ind(low_exposure['Wheezing'], high_exposure['Wheezing'])

print(f"T-statistic: {t_stat}")
print(f"P-value: {p_value}")

# Interpretation of p-value
if p_value < 0.05:
    print("There is a significant difference in Wheezing between the low and high exposure groups..")
else:
    print("No significant difference in Wheezing between the two groups.")
```
